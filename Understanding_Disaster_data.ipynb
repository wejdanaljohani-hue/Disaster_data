#Dataset Overview :

### ** Measures of Central Tendency**

```python
# Columns to analyze: Excluding CPI as it's not deemed useful for the analysis
central_tendency_features = ["Total Deaths", "No. Injured", "Total Damage ('000 US$)", "Disaster Group", "Event Name"]

# Initialize the stats dictionary
central_tendency_stats = {}

# For numeric columns, calculate mean and median
numeric_columns = ["Total Deaths", "No. Injured", "Total Damage ('000 US$)"]
central_tendency_stats['Mean'] = df[numeric_columns].mean()
central_tendency_stats['Median'] = df[numeric_columns].median()

# For categorical columns, calculate the mode
categorical_columns = ["Disaster Group", "Event Name"]
central_tendency_stats['Mode'] = {col: df[col].mode()[0] for col in categorical_columns}

# Convert the dictionary to a DataFrame for better display
central_tendency_stats_df = pd.DataFrame(central_tendency_stats)

# Display the results
central_tendency_stats_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Columns to visualize (numeric features)
central_tendency_features = ["Total Deaths", "No. Injured", "Total Damage ('000 US$)", "Magnitude"]
categorical_features = ["Disaster Group", "Event Name"]

# Check if the columns exist in the dataset
for feature in central_tendency_features + categorical_features:
    if feature not in df.columns:
        print(f"Warning: '{feature}' column is missing in the dataset!")

# Set up a 2-row, 2-column grid layout for histograms
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Central Tendency of Key Numerical Features (Mean, Median, Mode)", fontsize=16, fontweight='bold')

# Loop through features and create histograms with mean, median & mode lines for numeric columns
for i, feature in enumerate(central_tendency_features):
    row, col = divmod(i, 2)  # Calculate row and column index for subplot

    # Create a histogram
    sns.histplot(df[feature], bins=30, kde=True, ax=axes[row, col], color="skyblue")

    # Overlay mean, median, and mode lines
    axes[row, col].axvline(df[feature].mean(), color='red', linestyle='dashed', linewidth=2, label="Mean")
    axes[row, col].axvline(df[feature].median(), color='green', linestyle='dashed', linewidth=2, label="Median")
    axes[row, col].axvline(df[feature].mode().values[0], color='blue', linestyle='dashed', linewidth=2, label="Mode")

    # Labels and legend
    axes[row, col].set_title(feature.replace("_", " ").title(), fontsize=12, fontweight='bold')
    axes[row, col].set_xlabel("Value", fontsize=10)
    axes[row, col].legend()

# Loop through categorical features to calculate mode
for feature in categorical_features:
    mode_value = df[feature].mode()[0]
    print(f"Mode for '{feature}': {mode_value}")

# Adjust layout for better visualization
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### ** Measures of Dispersion**

```python
dispersion_features = ["Magnitude", "Total Affected", "Total Deaths", "Reconstruction Costs ('000 US$)"]

dispersion_stats = pd.DataFrame(index=dispersion_features)
dispersion_stats["Standard Deviation"] = df[dispersion_features].std()
dispersion_stats["Range"] = df[dispersion_features].max() - df[dispersion_features].min()
dispersion_stats["IQR"] = df[dispersion_features].quantile(0.75) - df[dispersion_features].quantile(0.25)

# Display the computed statistics
print("\n Measures of Dispersion")
display(dispersion_stats)
```
